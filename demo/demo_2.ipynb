```python
%matplotlib inline
import cv2
import matplotlib
import pylab as plt
import numpy as np
import math
import tensorflow as tf
import pandas as pd
import sys 

from tensorflow.keras.models import load_model
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras import regularizers
sys.path.append("../")
from models.config import get_default_configuration
from models.framing import get_frames
from models.model import get_model
from models.calculateDistance import calculateDistances
from models.detectparts import get_bodyparts
from set_label import set_label
from IPython.display import Video
```

```python
nFrame = get_frames("demo2.mp4", "demo2/")
```

```python
print(nFrame)
```

```python
df = get_bodyparts("demo2", nFrame )
```

Load the example data

```python
df.drop(["idFrame"], axis=1)
```

```python
model_3 = Sequential([  
    Dense(32, activation='relu', 
    kernel_regularizer=regularizers.l2(0.01),
    input_shape=(98,)),
    Dropout(0.3),
    Dense(16, activation='tanh',
    kernel_regularizer=regularizers.l2(0.01)), 
    Dropout(0.3),
    Dense(8,
    activation='tanh',
    kernel_regularizer=regularizers.l2(0.01)),    
    Dropout(0.3),    
    Dense(1, activation='sigmoid', 
    kernel_regularizer=regularizers.l2(0.01)),])
```

```python
model_3.compile(optimizer='adam',      
                loss='binary_crossentropy',    
                metrics=['accuracy'])
```

Load pre-trained weights

```python
model_3.load_weights('model_NN.h5')
```

```python
dataset = df.values
X = dataset[:,0:98]
```

Perform predictions on data

```python
predictions = model_3.predict_classes(X)
```
